```python
def get_avarage_pred(df1, df2, df3):
    all_preds = pd.concat([df1, df2, df3], axis=1)

    # Calculate the mean along the columns
    average_preds = all_preds.mean(axis=1)

    # Assign the result to a new column, e.g., "average_prediction"
    all_preds['average_prediction'] = average_preds

    # If you want to select only the average column
    return pd.DataFrame(all_preds['average_prediction']).rename(columns={"average_prediction": "pv_measurement"})
pred1 = get_avarage_pred(pred11, pred12, pred13)
pred2 = get_avarage_pred(perd21, pred22, pred23)
pred3 = get_avarage_pred(pred31, pred32, pred33)
pred11.head(20)
```
