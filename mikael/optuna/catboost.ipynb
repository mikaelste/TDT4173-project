```python
import pandas as pd
import numpy as np
from DataExplore import Data
```

```python
submission = f"catboost_optuna_{4}_kfold"
```

```python
data_A = Data("A")
```

```python
pred_a = data_A.catboost()
```

```python
pred_a = data_A.fit_predictions_to_submission_length(pred_a).to_numpy()
```

# B

```python
data_B = Data("B")
```

```python
pred_b = data_B.catboost()
```

```python
pred_b = data_B.fit_predictions_to_submission_length(pred_b).to_numpy()
```

# C

```python
data_C = Data("C")
```

```python
pred_c = data_C.catboost()
```

```python
df = pd.DataFrame({"prediction": prediction}).rename_axis(index="id")
df.to_csv(f"submissions/{submission}_submission.csv")
```
