```python
import pandas as pd
import numpy as np
from autogluon.tabular import TabularDataset, TabularPredictor
# from Data import Data
from DataExplore import Data
```

```python
submission = "auto_ml_submission_AUTOGLUON_"+str(5)
```

```python
data_a = Data("A")
data_b = Data("B")
data_c = Data("C")
```

```python
predictor_a = data_a.predict_location_GLUON()
```

```python
predictor_a.leaderboard()
```

```python
test_a = TabularDataset(data_a.test)
```

```python
y_pred_a = predictor_a.predict(test_a, as_pandas=False)
y_pred_a = data_a.fit_predictions_to_submission_length(y_pred_a).to_numpy()
```

## Pred B

```python
predictor_b = data_b.predict_location_GLUON()
```

```python
test_b = TabularDataset(data_b.test)
```

```python
y_pred_b = predictor_b.predict(test_b, as_pandas=False)
y_pred_b = data_b.fit_predictions_to_submission_length(y_pred_b).to_numpy()
```

## Pred C

```python
predictor_c = data_c.predict_location_GLUON()
```

```python
test_c = TabularDataset(data_c.test)
```

```python
y_pred_c = predictor_c.predict(test_c, as_pandas=False)
y_pred_c = data_c.fit_predictions_to_submission_length(y_pred_c).to_numpy()
```

## Submission

```python
prediction = list(y_pred_a) + list(y_pred_b) + list(y_pred_c)
```

```python
df = pd.DataFrame({"prediction": prediction}).rename_axis(index="id")
df.to_csv(f"submissions/{submission}_submission.csv")
```

# COMPARE BEST PREDICTION

```python
df_best = pd.read_csv("../submissions/predictions_auto.csv")
df_off = np.abs(df_best["prediction"] - df["prediction"])

A = df_off.copy().iloc[:720]
B = df_off.copy().iloc[720:720*2]
C = df_off.copy().iloc[(720*2):]
df_off_by_location = pd.DataFrame({"A":A.to_numpy(), "B":B.to_numpy(), "C":C.to_numpy()})
```

```python
df_off_by_location.describe()
```

```python
improved_mean = pd.read_csv("prev_submission.csv").mean() - df_off_by_location.mean()
improved_mean
```
